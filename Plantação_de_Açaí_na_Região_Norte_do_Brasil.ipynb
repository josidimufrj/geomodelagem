````python
df = xr.open_dataset('/content/drive/MyDrive/Colab Notebooks/vegetação_belem_acai.nc',engine='h5netcdf')
df.to_dataframe().reset_index()
````

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Select a variable to plot and the first time step
# For example, 'cvh' (high_vegetation_cover)
data_to_plot = df['cvh'].isel(valid_time=0) # Corrected: Selecting the first time step using 'valid_time'

# Get latitude and longitude from the dataset
latitude = df['latitude'].values
longitude = df['longitude'].values

# Define the projection
projection = ccrs.PlateCarree()

plt.figure(figsize=(10, 8))
ax = plt.axes(projection=projection)

# Plot the data
c = ax.pcolormesh(longitude, latitude, data_to_plot, transform=ccrs.PlateCarree(), cmap='viridis')

# Add geographical features
ax.coastlines(resolution='10m', color='black', linewidth=1)
ax.add_feature(cartopy.feature.BORDERS, linestyle=':', edgecolor='black')
ax.add_feature(cartopy.feature.STATES, linestyle=':', edgecolor='gray')

# Add gridlines
gridlines = ax.gridlines(draw_labels=True, linestyle='--', color='gray', alpha=0.5)
gridlines.top_labels = False
gridlines.right_labels = False

# Set plot title and colorbar
plt.colorbar(c, ax=ax, orientation='vertical', label=data_to_plot.long_name)
plt.title(data_to_plot.long_name )
ax.set_title(data_to_plot.long_name )

plt.show()
````

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

nome_var = "cvh"
print(type(df))
# Média mensal por ponto da grade
variavel = (
    df.groupby(["latitude", "longitude"])[nome_var]
      .mean()
      .reset_index()
)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES)

scatter = ax.scatter(
    variavel["longitude"],
    variavel["latitude"],
    c=variavel[nome_var],
    cmap="YlGn",
    s=250,
    edgecolor="black",
    transform=ccrs.PlateCarree()
)

ax.set_extent([
    variavel["longitude"].min()-0.1,
    variavel["longitude"].max()+0.1,
    variavel["latitude"].min()-0.1,
    variavel["latitude"].max()+0.1
])

gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False

plt.colorbar(scatter, label=variavel.columns[2])
plt.title(variavel.columns[2] )

plt.show()
````

````python
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import cartopy.crs as ccrs

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection=ccrs.PlateCarree())

# 🌍 CONTORNO DE TERRA (continentes)
ax.add_feature(cfeature.LAND, facecolor="lightgray")

# 🌊 oceano (opcional)
ax.add_feature(cfeature.OCEAN, facecolor="lightblue")

# 🌎 fronteiras e costas
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=1)

# 📍 área
ax.set_extent([-52, -46, -4, 0], crs=ccrs.PlateCarree())

plt.title("Mapa com contorno de terra")
plt.show()
````
